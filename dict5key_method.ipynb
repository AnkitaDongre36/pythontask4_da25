```python
data = {
    "id": 101,                     
    "name": "Ankita",              
    "skills": ["Python", "SQL"],   
    "address": {                   
        "city": "Nashik",
        "state": "Maharashtra"
    },
    "experience": 3              
}

print(data)
```

### get()	
** Returns the value of the specified key **

```python
x=data.get("address")
print(x)
```

### pop()	
** Removes the element with the specified key **

```python
data.pop("name")
print(data)
```

### clear()	
** Removes all the elements from the dictionary **

```python
data.clear()
print(f"data={data}")
```

```python
animal = {
    "id": 1,                              
    "name": "Lion",                        
    "diet": "Carnivore",                   
    "features": ["Strong", "Fast"],        
    "habitat": {"place": "Forest", "country": "India"}  
}
print(f"animal={animal},len={len(animal)},type={type(animal)}")
```

### get()	
** Returns the value of the specified key **

```python
x=animal.get("diet")
print(x)
```

### Append

```python
animal["features"].append("powerful body")
print(animal["features"])                         
```

### values
** Returns a list of all the values in the dictionary **

```python
x=animal.values()
print(x)
```

```python
fruit = {
    "id": 10,                                   
    "name": "Mango",                            
    "color": "Yellow",                       
    "vitamins": ["A", "C"],                    
    "details": {"taste": "Sweet", "country": "India"}   
}
print(f"fruit={fruit},type={type(fruit)}")
```

### keys()	
** Returns a list containing the dictionary's keys **

```python
x=fruit.keys()
print(x)
```

### copy()	
** Returns a copy of the dictionary **

```python
y=fruit.copy()
print(y)
```

#### values()	
** Returns a list of all the values in the dictionary **

```python
x=fruit.values()
print(x)
```

```python
dress = {
    "id": 501,                                
    "name": "Saree",                          
    "color": "Red",                           
    "sizes": ["S", "M", "L"],                 
    "details": {"material": "Silk", "brand": "FabIndia"}   
}
print(f"dress={dress}")
```

### keys()

```python
x=dress.keys()
print(x)
```

### update()

```python
z=dress.values()
print(z)
```
