```python
animal = {
    "name": "Tiger",                 
    "scientific_name": "Panthera tigris",   
    "age": 5,                        
    "weight": 220,                   
    "color": "Orange with black stripes",  
    "habitat": "Forest",             
    "is_carnivore": True,           
    "speed_kmph": 60,               
    "lifespan": 15,                  
    
    "foods": ["Deer", "Wild boar", "Buffalo"],   
    "countries_found": ["India", "Nepal", "Bangladesh"],  
    "body_parts": ["Head", "Tail", "Legs", "Claws"],      
    
    "population": {                  
        "india": 3000,
        "world": 3900
    },
    
    "classification": {              
        "kingdom": "Animalia",
        "class": "Mammalia",
        "family": "Felidae"
    },
    
    "features": ["Strong", "Fast", "Sharp teeth"]  
}

print(f"animal={animal}")
```

### values()	
** Returns a list of all the values in the dictionary **

```python
animalfeature=animal.values()
print(animalfeature)
```

### pop()	
** Removes the element with the specified key **

```python
animal.pop("classification")
print(f"animal={animal}")
```

### update()	
** Updates the dictionary with the specified key-value pairs **

```python
animal.update({"classification":{"kingdom": "Animalia",
        "class": "Mammalia",
        "family": "Felidae"}})
print(animal)
```

```python
student = {
    "id": 101,                         
    "name": "Ankita",                  
    "age": 22,                         
    "gender": "Female",                
    "course": "Data Analytics",        
    "year": 3,                         
    "college": "ABC College",          
    "city": "Pune",                    
    
    "subjects": ["Python", "SQL", "Statistics"],   
    "skills": ["Excel", "Power BI", "Python"],     
    "hobbies": ["Reading", "Drawing", "Music"],    
    
    "marks": {                         
        "Python": 85,
        "SQL": 90,
        "Statistics": 88
    },
    
    "address": {                       
        "area": "Chinchwad",
        "state": "Maharashtra"
    },
    
    "attendance": 92,                  
    "is_passed": True                  
}
print(f"student={student}")
```

### copy()	
** Returns a copy of the dictionary **

```python
studentinfo=student.copy()
print(f"studentinfo={studentinfo}")
```

### keys()	
** Returns a list containing the dictionary's keys**

```python
studentkeys=studentinfo.keys()
print(f"studentkeys={studentkeys}")
```

### get()	
**  Returns the value of the specified key **

```python
datareturn=studentinfo.get("subjects")
print(f"datareturn={datareturn}")
```
