```python
student = {
    "name": "Neha",
      "age": 23
      }
print(f"student={student}")
```

### update()	
** Updates the dictionary with the specified key-value pairs **

```python
student.update({"address":"Nagpur"})
print(student)
```

### keys()	
** Returns a list containing the dictionary's keys **

```python
x=student.keys()
print(x)
```

### values()	
** Returns a list of all the values in the dictionary **

```python
y=student.values()
print(y)
```

```python
fruit = {
    "name": "Mango",
    "color": "Yellow",
    "vitamins": ["A", "C", "E"]
}
print(f"fruit={fruit}")
```

### Append
** Add the value of last index **

```python
fruit["vitamins"].append("D")
print(fruit["vitamins"])
```

### pop()	
** Removes the element with the specified key **

```python
fruit.pop("color")
print(fruit)
```

### update()	
** Updates the dictionary with the specified key-value pairs **

```python
fruit.update({"color":"yellow"})
print(fruit)
```

```python
animal = {
    "id": 1,
    "name": "Tiger",
    "type": "Wild",
    "habitat": {
        "place": "Forest",
        "country": "India"
    }
}

print(f"animal={animal},type={type(animal)}")
```

### values()	
** Returns a list of all the values in the dictionary **

```python
x=animal.values()
print(x)
```

### pop()	
** Removes the element with the specified key **

```python
animal["habitat"].pop("place")
print(f"animal={animal}")
```

### update()	
** Updates the dictionary with the specified key-value pairs **

```python
animal["habitat"].update({"place":"Forest"})
print(f"animal={animal}")
```

```python
hotel = {
    "id": 201,
    "name": "Taj Hotel",
    "city": "Mumbai",
    "rooms": ["Single", "Double", "Suite"],
    "manager": {
        "name": "Rahul",
        "experience": 10
    }
}
print(f"hotel={hotel}")
```

### values()	
** Returns a list of all the values in the dictionary **

```python
x=hotel.values()
print(x)
```

### copy()	
** Returns a copy of the dictionary **

```python
hotelnagpur=hotel.copy()
print(f"hotelnagpur={hotelnagpur}")
```

### pop()	
** Removes the element with the specified key **

```python
hotelnagpur.pop("city")
print(f"hotelnagpur={hotelnagpur}")
```
